## Step 5: Understand the Flow

Every ClawBio skill follows the same pattern:

| Component | Role |
|-----------|------|
| **SKILL.md** | Contract: inputs, outputs, domain decisions, safety rules |
| **Python script** | Implementation: reads data, runs pipeline, writes report |
| **Demo data** | Synthetic test data shipped with every skill |
| **AI agent** | (Optional) dispatches the skill, passes parameters, explains results |

The agent reads the SKILL.md to understand the skill. It **never overrides** domain decisions or safety rules.
